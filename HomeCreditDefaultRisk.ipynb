## Codificando categorias

```python
categorical_features = [i for i in home_train.columns if home_train.dtypes[i] in ['object', 'category']]
print(categorical_features)
```

```python
categorical_names = {}
encoders = {}

for feature in categorical_features:
  le = LabelEncoder()
  le.fit(home_train[feature])

  home_train[feature] = le.transform(home_train[feature])

  categorical_names[feature] = le.classes_
  encoders[feature] = le
```

Limpando Valores Nulos  Objetivo



```python
home_train_SemNulos= home_train.dropna()
print(home_train.shape)
print(home_train_SemNulos.shape)

ht = home_train_SemNulos
```

```python
print(ht['Class'].value_counts())
print(home_train['Class'].value_counts())
```

```python
ht = ht.astype(int)
ht.dtypes
# todos devem estar codificados em int64
```

```python
# Definições sobre os dados para avaliação de justiça
ht.at[71, 'Class'] = 0
ht.at[124, 'Class'] = 1
label_name ='Class'
protected_attribute_names = ['Sexo', 'faixa_etaria']
favorable_classes = np.where(ht['Class'] == 0)[0]
```

```python
print(favorable_classes)
print(ht.head())
```

```python
# Definindo as populações privilegiadas e não-privilegiadas
# Sexo
## Recuperando o indice
privileged_sex   = np.where(categorical_names['Sexo'] == 'M')[0]
unprivileged_sex = np.where(categorical_names['Sexo'] == 'F')[0]
## Estabelecendo formato, seguindo o formato da documentacao das classes ClassificationMetric and BinaryLabelDatasetMetric
privileged_groups   = [{'Sexo' : privileged_sex}]
unprivileged_groups = [{'Sexo' : unprivileged_sex}]

# Idade
## Recuperando o indice
privileged_faixa_etaria   = np.where(categorical_names['faixa_etaria'] == 'adulto')[0]
unprivileged_faixa_etaria = np.where(categorical_names['faixa_etaria'] == 'jovem')[0]
## Estabelecendo formato, seguindo o formato da documentacao das classes ClassificationMetric and BinaryLabelDatasetMetric
privileged_groups_fe   = [{'faixa_etaria' : privileged_faixa_etaria}]
unprivileged_groups_fe = [{'faixa_etaria' : unprivileged_faixa_etaria}]


# Dados privilegiados
privileged_classes = [privileged_sex, privileged_faixa_etaria]
```
