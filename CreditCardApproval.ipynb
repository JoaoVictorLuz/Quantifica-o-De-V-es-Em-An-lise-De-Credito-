## Codificando categorias

```python
categorical_features = [i for i in CreditCard.columns if CreditCard.dtypes[i] in ['object', 'category']]
print(categorical_features)
```

```python
categorical_names = {}
encoders = {}

for feature in categorical_features:
  le = LabelEncoder()
  le.fit(CreditCard[feature])

  CreditCard[feature] = le.transform(CreditCard[feature])

  categorical_names[feature] = le.classes_
  encoders[feature] = le
```

```python
print(len(CreditCard))
CreditCard = CreditCard.dropna()
print(len(CreditCard))
CreditCard = CreditCard.astype(int)
# todos devem estar codificados em int64
CreditCard.dtypes
```

```python
print(CreditCard.head())
```

```python
# Definições sobre os dados para avaliação de justiça
CreditCard.at[1, 'Class'] = 0
CreditCard.at[0, 'Class'] = 1
label_name ='Class'
protected_attribute_names = ['Gender', 'Education']
favorable_classes = np.where(CreditCard['Class'] == 0)[0]
```

```python
# Definindo as populações privilegiadas e não-privilegiadas
# Gênero
## Recuperando o indice
privileged_sex   = np.where(categorical_names['Gender'] == 'M')[0]
unprivileged_sex = np.where(categorical_names['Gender'] == 'F')[0]
## Estabelecendo formato, seguindo o formato da documentacao das classes ClassificationMetric and BinaryLabelDatasetMetric
privileged_groups   = [{'Gender' : privileged_sex}]
unprivileged_groups = [{'Gender' : unprivileged_sex}]

#Escolaridade
## Recuperando o indice
privileged_faixa_etaria   = np.where(categorical_names['Education'] == 'Graduado')[0]
unprivileged_faixa_etaria = np.where(categorical_names['Education'] == 'NaoGraduado')[0]
## Estabelecendo formato, seguindo o formato da documentacao das classes ClassificationMetric and BinaryLabelDatasetMetric
privileged_groups_fe   = [{'Education' : privileged_faixa_etaria}]
unprivileged_groups_fe = [{'Education' : unprivileged_faixa_etaria}]


# Dados privilegiados
privileged_classes = [privileged_sex, privileged_faixa_etaria]
```
